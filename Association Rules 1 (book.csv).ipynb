```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_csv("C:\\Users\\pradyumn\\Documents\\Assignment\\book.csv")
df.head(5)
```

# Apriori Algorithm

```python
set = apriori(df, min_support=0.1, use_colnames=True)
set
```

```python
set.shape
```

```python
rules = association_rules(set, metric="lift", min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending = False).head(10)
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
sns.scatterplot(x='support',y='confidence', data= rules)
```

```python
set1 = apriori(df, min_support=0.2, use_colnames=True)
set1
```

```python
rules1 = association_rules(set1, metric="lift", min_threshold=1)
rules1
```

```python
rules1.sort_values('lift',ascending = False).head(10)
```

```python
rules1.sort_values('lift',ascending = False)[0:20]
```

```python
rules1[rules1.lift>1]
```

```python
sns.scatterplot(x='confidence',y='support',data=rules1)
```
